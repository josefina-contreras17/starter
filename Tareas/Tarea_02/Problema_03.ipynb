```python
primer_puntaje = int(input("Ingresa primer puntaje: "))
segundo_puntaje = int(input("Ingresa segundo puntaje: "))

suma = primer_puntaje + segundo_puntaje

if suma < 50:
    print("Resultado: Nivel principiante")
elif 50 <= suma < 80:
    print("Resultado: Nivel intermedio")
else:
    print("Resultado: Nivel pro")
```
