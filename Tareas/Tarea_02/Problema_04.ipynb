```python
def calcular_perfil_musical(edad, le_gusta_musica, le_gusta_bailar):
  """
  Args:
    edad (int): La edad del usuario.
    le_gusta_musica (bool): True si le gusta la música, False en caso contrario.
    le_gusta_bailar (bool): True si le gusta bailar, False en caso contrario.

  Returns:
    str: El mensaje del perfil musical del usuario.
  """
  if edad >= 18 and le_gusta_musica and le_gusta_bailar:
    return "Eres una estrella de la fiesta"
  elif le_gusta_musica and not le_gusta_bailar:
    return "Te gusta la música, pero no tanto bailar"
  else:
    return "Tienes tu propio estilo"
```

### Ejemplos de uso:

```python
# Ejemplo 1: Una estrella de la fiesta (Edad >= 18, le gusta la música, le gusta bailar)
perfil1 = calcular_perfil_musical(25, True, True)
print(f"Perfil 1: {perfil1}")

# Ejemplo 2: Te gusta la música, pero no tanto bailar (le gusta la música, no le gusta bailar)
perfil2 = calcular_perfil_musical(30, True, False)
print(f"Perfil 2: {perfil2}")

# Ejemplo 3: Tienes tu propio estilo (menor de 18, aunque le guste todo)
perfil3 = calcular_perfil_musical(16, True, True)
print(f"Perfil 3: {perfil3}")

# Ejemplo 4: Tienes tu propio estilo (no le gusta la música)
perfil4 = calcular_perfil_musical(22, False, True)
print(f"Perfil 4: {perfil4}")

# Ejemplo 5: Tienes tu propio estilo (no le gusta nada)
perfil5 = calcular_perfil_musical(40, False, False)
print(f"Perfil 5: {perfil5}")

# Ejemplo 6: Tienes tu propio estilo (le gusta bailar pero no la música)
perfil6 = calcular_perfil_musical(28, False, True)
print(f"Perfil 6: {perfil6}")
```
